```python
# Import Df 
df_ords = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'orders_checked.csv'))
```

```python
# Import new df 
df_ords_prior = pd.read_csv(os.path.join(path, '02 Data', 'Original Data', 'order_products__prior.csv'), index_col = False)
```

```python
# Drop Extra index Columns
df_ords = df_ords.drop(columns = ['Unnamed: 0.1', 'Unnamed: 0'])
```

```python
# Check Output
df_ords
```

```python
# Check dimensions of df_ords_prior 
df_ords_prior.shape
```

```python
# Test the merge 
pd.merge(df_ords, df_ords_prior, on = ['order_id'], indicator = True)
```

```python
# Merge dataframes 
df_merged_large = df_ords.merge(df_ords_prior, on = 'order_id', indicator = True)
```

```python
# Check output 
df_merged_large.head()
```

```python
df_merged_large['_merge'].value_counts()
```

```python
# Create an outer merge
df_merged_large_2 = df_ords.merge(df_ords_prior, on = 'order_id', how = 'outer', indicator = True)
```

```python
#Check value counts for outer merge
df_merged_large_2['_merge'].value_counts()
```

```python
# Drop added merge column 
df_merged_large = df_merged_large.drop(columns = ['_merge'])
```

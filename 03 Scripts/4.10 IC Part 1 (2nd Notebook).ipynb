### Create column based on age
Young Adult <35 Middle Aged Adult 35-64 Senior >=65

```python
df_new.loc[df_new['age'] < 35, 'age_flag'] = 'Young Adult' 
df_new.loc[(df_new['age'] >= 35) & (df_new['age'] < 65), 'age_flag'] = 'Middle-Aged Adult'
df_new.loc[df_new['age'] >= 65, 'age_flag'] = 'Senior Adult'
```

```python
# Check value count for new column 
# The majority of customers are middle_aged adults
df_new['age_flag'].value_counts(dropna = False)
```

```python
df_new['income'].describe()
```

### The minimum income is 25,903 and the max is 593,901, with a mean of 99,675 (USD)

#### Create income column: 
Low Income: less than 40,000 
Working Class: 40,000 - 90,000 
Middle Class: 90,000 - 175,000 
Upper Class: More than 200,000

```python
df_new.loc[df_new['income'] < 40000, 'income_class'] = 'Low Income' 
df_new.loc[(df_new['income'] >= 40000) & (df_new['income'] < 90000), 'income_class'] = 'Working Class'
df_new.loc[(df_new['income'] >= 90000) & (df_new['income'] < 175000), 'income_class'] = 'Middle Class'
df_new.loc[df_new['income'] >= 175000, 'income_class'] = 'Upper Class'
```

```python
# Check value counts 
df_new['income_class'].value_counts(dropna = False)
```

```python
# Compare age flag to income flag 
pd.crosstab(df_new['age_flag'], df_new['income_class'], dropna = False)
```

#### Most middle-aged and senior adults are in the middle class, while young adults are in the working class

#### Create flag for family size 

No children: 0 dependents 
Small family: 1-2 children
Large Family: More than 2 children

```python
df_new.loc[df_new['#_dependents'] < 1, 'family_size'] = 'No Children'
df_new.loc[(df_new['#_dependents'] >= 1) & (df_new['#_dependents'] <=2), 'family_size'] = 'Small Family'
df_new.loc[df_new['#_dependents'] > 2, 'family_size'] = 'Large Family'
```

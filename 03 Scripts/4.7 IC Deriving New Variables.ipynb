```python
# Define function 

def price_label(row):
    if row['prices'] <= 5: 
        return 'Low-Range Product'
    elif (row['prices'] > 5) and (row['prices'] <= 15): 
        return 'Mid-Range Product' 
    elif row['prices'] > 15: 
        return 'High-Range Product'
    else: return 'Not enough data'
```

## 04. Identify busiest hours of the day and create new column with "Most Orders, Fewest Orders, Average Orders"

```python
 # Step 1: Find busiest hours 

prods_ords_merged['order_hour'].value_counts(dropna = False)
```

```python
# Create if-for statement defining busiest hours as > 2500000, least busy hours as <100000 and regularly busy hours (all hours in between)

hours = [] 

for value in prods_ords_merged['order_hour']: 
    if value == 10 or value == 11 or value == 14 or value == 15 or value == 13 or value == 12 or value == 16: 
        hours.append('Most Orders') 
    elif value == 5 or value == 2 or value == 4 or value == 3: 
        hours.append('Fewest Orders')
    else: 
        hours.append('Average Orders')
```

```python
hours
```

```python
# 04: Add list to new column 'busiest_period_of_day'
prods_ords_merged['busiest_period_of_day'] = hours
```
